# Recebendo 2 valores através do input


```python
num1 = input("Digite um numero: ")
print(num1)
print(type(num1))
num2 = input("Digite outro numero: ")
print(num2)
print(type(num2))
```

# Convertendo os numeros para ponto flutuante

```python
num1 = float(num1)
print(type(num1))
num2 = float(num2)
print(type(num2))
```

# Aplicando 4 operações matemáticas no código

```python
#Adicação
soma = (num1 + num2)
print(soma)
```

```python
#subtração
subtracao = num1 - num2
print(subtracao)
```

```python
#multiplicação
multiplicacao = num1 * num2
print(multiplicacao)
```

```python
#divisão
divisao = num1 / num2
print(divisao)
```

# Adicionando um laço de repetição e permitindo que o usuário realize a operação matemática desejada:

```python
resp = "s" #força a primeira entrada no loop com a reposta sim

while (resp == "s"):
  num1 = float(input("Digite um numero: "))
  num2 = float(input("Digite outro numero: "))


  print("Vamos calcular: ")

  print("Digite a operação desejada:")
  print("1 - Adição")
  print("2 - Subtração")
  print("3 - Multiplicação")
  print("4 - Divisão")

  operacao = int(input())

  if(operacao == 1):
    soma = num1 + num2
    print("O resultado da sua soma é:",soma)

  elif(operacao == 2):
    subtracao = num1 - num2
    print("O resultado da sua subtração é:",subtracao)

  elif(operacao == 3):
    multiplicacao = num1 * num2
    print("O resultado da sua multiplicação é: ",multiplicacao)

  elif(operacao == 4):
    divisao = num1 / num2
    print("O resultado da sua divisãoé: ",divisao)

  else:
    print("Operação inválida")

  print("Deseja continuar calculando: [S/N]")
  resp = input().lower() #converte a resposta para minusculo em caso do usuário digitar maiúsculo
```
